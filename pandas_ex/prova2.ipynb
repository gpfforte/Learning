```python
import pandas as pd

# Creazione di un DataFrame
data = {'Nome': ['Alice', 'Bob', 'Charlie'],
        'Età': [25, 30, 22],
        'Città': ['Roma', 'Milano', 'Napoli']}
df_prova = pd.DataFrame(data)

# Selezione di una colonna
print(df_prova['Età'])

# Filtraggio dei dati
print(df_prova[df_prova['Età'] > 25])

# Raggruppamento e aggregazione
print(df_prova.groupby('Città')['Età'].mean())

# Creazione di un grafico
df_prova.plot(kind='bar', x='Nome', y='Età')
```
