```python
import pandas as pd
url = "https://it.wikipedia.org/wiki/Prefissi_telefonici_italiani"
df = pd.read_html(url)[0]
print(type(df.columns))
print(df.columns)
converters = {c:lambda c: str(c) for c in df.columns}
# print(converters)
# converters = {i:str for i in range(len(df.columns))}
print(converters)
print(len(df.columns))
df = pd.read_html(url, converters=converters)[0]


df.to_excel("prefissi_italiani.xlsx", index=False)
df
```
