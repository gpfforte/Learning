```python
df1 = pd.DataFrame({"a": [1, 0, 1], "b": [0, 1, 1]}, dtype=bool)
df1
```

```python
df2 = pd.DataFrame({"a": [0, 1, 1], "b": [1, 1, 0]}, dtype=bool)
df2
```

```python
# XOR
df1 ^ df2
```

```python
# AND
df1 & df2
```

```python
# OR
df1 | df2
```
