```python
import pandas as pd
from datetime import datetime as dt

now = dt.now()
date_time = now.strftime("%Y%m%d-%H%M%S")
date_time
```

```python
df = pd.read_excel("CARLI AIRASCA 2024-RCL-000894 10.12.2024.xlsx",skiprows=1)
df.head()
```

```python
df_new = df.rename(columns=str.upper)
df_new.columns = df_new.columns.str.replace(' ', '_')
df_new["BARCODE"] = df_new["DATA_ENTRATA"].dt.strftime("%Y%m%d")+ "|" + df_new["N°_BOLLA"].astype(str)
print(len(df_new))
df_new.head()
```

```python
df_new[["N°_BOLLA", "DATA_ENTRATA", "DESTINATARIO", "BARCODE"]].to_csv(f"Cirillo_Airasca_{date_time}.csv",sep=";", index=False)
```
