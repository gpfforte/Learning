```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
np.set_printoptions(8)
arr_x = np.linspace(-3*np.pi, 3*np.pi, 1000)
arr_y = np.sin(arr_x)
arr_z = np.cos(arr_x)
plt.style.use("bmh")
plt.plot(arr_x,arr_y, label="Seno")
plt.plot(arr_x,arr_z, label="Coseno")
plt.legend()
plt.tight_layout()
```

<h2>
Soluzione Sistemi Lineari (rette)
</h2>

```python
# Matrice Coefficienti
coef = np.array([[2,-1],[1,3]])

print("coef:\n", coef)
# Vettore delle costanti
ordi = np.array([4,9])
print("ordi:\n", ordi)

# Nel caso di rette posso disegnare le rette nel piano
# Di seguito le due funzioni
def f0(x):
    return ((+ordi[0]-coef[0,0]*x)/coef[0,1])
def f1(x):
    return ((+ordi[1]-coef[1,0]*x)/coef[1,1])

# risolvo anticipatamente il sistema cos disegno le rette nei dintorni dell'intersezione
x = np.linalg.solve(coef, ordi)

xx=np.linspace(x[0]-2.5, x[0]+2.5, 100)
f0y=f0(xx)
f1y=f1(xx)
plt.style.use("bmh")
plt.plot(xx,f0y, label = "f0y")
plt.plot(xx,f1y, label = "f1y")
plt.legend()

# Sistema pi ottimizzato di risoluzione che vale anche per n dimensioni
x = np.linalg.solve(coef, ordi)
print("x:", x)
print(f"x = {x[0]:.01f}", f"y = {x[1]:.01f}")

# Sistema meno ottimizzato di risoluzione con calcolo matriciale vale comunque per n dimensioni
y = np.linalg.inv(coef).dot(ordi)
print(f"x = {y[0]:.01f}", f"y = {y[1]:.01f}")

# Sistema con il metodo di Cramer o dei determinanti
coef_x = coef.copy()
coef_y = coef.copy()
# Ora inserisco le costanti nella matrice dei coeffienti al posto dei coefficienti sia delle x che delle y
# Questo vale solo per 2 dimensioni, per farlo per pi dimensioni occorre aggiungere i calcoli anche per 
# le dimensioni in pi
coef_x[:,0] = ordi
coef_y[:,1] = ordi
print("coef_x:\n", coef_x)
print("coef_y:\n", coef_y)
det_coef = round(np.linalg.det(coef), 5)
print("det_coef:\n", det_coef)
det_x = round(np.linalg.det(coef_x), 5)
print("det_x:", det_x)
det_y = round(np.linalg.det(coef_y), 5)
print("det_y:", det_y)

x = det_x/det_coef
y = det_y/det_coef

print(f"x = {x}",f"y = {y}")
```
